```python
import pandas as pd
import matplotlib.pyplot as plt
import json

# Load the JSON file
with open('data4.json', 'r') as file:
    data = json.load(file)

# Load data into Pandas DataFrames
departments_df = pd.DataFrame(data['departments'])
levels_df = pd.DataFrame(data['levels'])

# Merge levels with departments to include department names
levels_df = levels_df.merge(departments_df, left_on='department_id', right_on='id', suffixes=('_level', '_department'))

# Visualization 1: Total students per department
students_per_department = levels_df.groupby('name')['students'].sum()

plt.figure(figsize=(10, 6))
students_per_department.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Students per Department')
plt.xlabel('Department')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("students_per_department.png")
plt.show()

# Visualization 2: Students per level for Software Engineering
software_eng_levels = levels_df[levels_df['name'] == 'Software Engineering']

plt.figure(figsize=(10, 6))
plt.plot(software_eng_levels['level'], software_eng_levels['students'], marker='o', linestyle='-', color='green')
plt.title('Students per Level in Software Engineering')
plt.xlabel('Level')
plt.ylabel('Number of Students')
plt.grid()
plt.tight_layout()
plt.savefig("students_per_level.png")
plt.show()
```
